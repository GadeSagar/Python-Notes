# Seaborn Charts using Cars93 Dataset

- Seaborn is a Python data visualization library built on top of Matplotlib.
It provides a high-level interface for creating statistical graphics that are informative, attractive, and easy to interpret.

- Seaborn is specifically designed to work well with Pandas DataFrames and is widely used in Exploratory Data Analysis (EDA) and Data Science projects.

### Why Seaborn is Used (Problem it Solves)

Matplotlib is powerful but:
- Requires more code
- Needs manual styling
- Does not handle statistical plots easily

**Seaborn solves this by:**
- Providing default themes
- Automatically computing statistics (mean, CI, distribution)
- Making EDA faster and cleaner

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

sns.set_theme(style="whitegrid")   #Clean visual styles without extra code

df = pd.read_csv("C:/Users/sagar/Desktop/All Data/Data Sci/csv files/cars93.csv")
df.head()
```

## 1. Scatter Plot
**Definition:** Shows relationship between two continuous variables.

**Use Case:** Identify correlation, trends, and outliers.

```python

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="EngineSize", y="Price", hue="Origin", size="Horsepower", palette="Set2")
plt.title("Engine Size vs Price")
plt.show()
```

## 2. Line Plot
**Definition:** Shows trend over ordered data.

**Use Case:** Trend analysis.

```python

df_sorted = df.sort_values("EngineSize")
plt.figure(figsize=(8,5))
sns.lineplot(data=df_sorted, x="EngineSize", y="MPG.highway", marker="o", ci=None)
plt.title("Highway MPG Trend by Engine Size")
plt.show()
```

## 3. Bar Plot
**Definition:** Shows aggregated statistics.

**Use Case:** Compare categories.

```python

plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Type", y="Price", palette="coolwarm")
plt.title("Average Price by Car Type")
plt.show()
```

## 4. Count Plot
**Definition:** Shows frequency of categories.

```python

plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Origin", palette="pastel")
plt.title("Number of Cars by Origin")
plt.show()
```

## 5. Histogram
**Definition:** Shows distribution of a numeric variable.

```python

plt.figure(figsize=(7,4))
sns.histplot(df["Price"], bins=15, kde=True)
plt.title("Price Distribution")
plt.show()
```

## 6. Box Plot
**Definition:** Shows quartiles and outliers.

```python

plt.figure(figsize=(7,5))
sns.boxplot(data=df, x="Origin", y="Price", palette="Set3")
plt.title("Price Distribution by Origin")
plt.show()
```

## 7. Violin Plot
**Definition:** Shows distribution shape.

```python

plt.figure(figsize=(7,5))
sns.violinplot(data=df, x="Origin", y="MPG.city", inner="quart", palette="muted")
plt.title("City MPG Distribution by Origin")
plt.show()
```

## 8. Pair Plot
**Definition:** Shows pairwise relationships.

```python

cols = ["Price", "EngineSize", "Horsepower", "MPG.city"]
sns.pairplot(df[cols], diag_kind="kde")
plt.show()
```

## 9. Heatmap
**Definition:** Shows correlation matrix.

```python

corr = df[["Price", "EngineSize", "Horsepower", "Weight", "MPG.city"]].corr()
plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, cmap="vlag", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

## 10. Regression Plot
**Definition:** Shows linear relationship.

```python

plt.figure(figsize=(7,5))
sns.regplot(data=df, x="Horsepower", y="Price")
plt.title("Horsepower vs Price (Regression)")
plt.show()
```

## 11. relplot 
**Definition:** relplot is a figure-level function used to show relationships between numerical variables.
It supports both scatter and line plots and allows faceting.

```python
sns.relplot(
    data=df,
    x="EngineSize",
    y="Price",
    hue="Origin",
    style="Type",
    kind="scatter",
    height=5,
    aspect=1.2
)
plt.suptitle("Relational Plot: Engine Size vs Price", y=1.02)
plt.show()
```

## 12. stripplot
**Definition:** stripplot shows individual data points for a categorical variable with optional jitter.
- Small datasets
- Show exact values without aggregation

```python
plt.figure(figsize=(8,5))
sns.stripplot(
    data=df,
    x="Origin",
    y="Price",
    jitter=True,
    alpha=0.7
)
plt.title("Strip Plot: Price by Origin")
plt.show()
```

## 13. residplot
**Definition:** residplot displays residuals of a regression model to check model assumptions.
- Validate linear regression
- Detect non-linearity and heteroscedasticity

```python
plt.figure(figsize=(7,4))
sns.residplot(
    data=df,
    x="Horsepower",
    y="Price",
    lowess=True
)
plt.title("Residual Plot: Horsepower vs Price")
plt.show()
```

## 14. swarmplot
**Definition:** swarmplot is similar to stripplot but avoids overlapping points.
- Precise distribution visualization for small-to-medium datasets

```python
plt.figure(figsize=(8,5))
sns.swarmplot(
    data=df,
    x="Origin",
    y="MPG.city",
    hue="Origin"
)
plt.title("Swarm Plot: City MPG by Origin")
plt.show()
```

## 15. kdeplot
**Definition:** kdeplot shows a smooth probability density curve of a numeric variable.
- Understand distribution shape
- Compare multiple distributions

```python
plt.figure(figsize=(7,4))
sns.kdeplot(
    data=df,
    x="Price",
    hue="Origin",
    fill=True,
    common_norm=False
)
plt.title("KDE Plot: Price Distribution by Origin")
plt.show()
```

## 16. displot
**Definition:** displot is a figure-level distribution function that can draw histograms, KDEs, or ECDFs.
- Distribution analysis with faceting

```python
sns.displot(
    data=df,
    x="MPG.highway",
    col="Origin",
    kde=True,
    bins=12,
    height=4,
    aspect=0.9
)
plt.suptitle("Distribution of Highway MPG by Origin", y=1.05)
plt.show()
```

## 17.rugplot
**Definition:** rugplot displays small ticks for each observation along an axis.
- Show raw data density
- Often used with KDE

```python
plt.figure(figsize=(7,2))
sns.rugplot(
    data=df,
    x="Price",
    height=0.5
)
plt.title("Rug Plot: Price Values")
plt.show()
```

## 18. regplot (Regression Plot)
**Definition:** regplot shows scatter data + regression line with confidence interval..
- Relationship validation
- Linear trend analysis

```python
plt.figure(figsize=(7,5))
sns.regplot(
    data=df,
    x="EngineSize",
    y="Price",
    scatter_kws={"alpha":0.6}
)
plt.title("Regression Plot: Engine Size vs Price")
plt.show()
```

## 19.catplot (Categorical Plot)
**Definition:** catplot is a figure-level categorical plotting function supporting multiple plot types.
- Faceted categorical analysis

```python
sns.catplot(
    data=df,
    x="Origin",
    y="Price",
    kind="box",
    col="Type",
    height=4,
    aspect=0.8
)
plt.suptitle("Catplot: Price by Origin and Type", y=1.05)
plt.show()
```

## 20. lmplot (Regression + Facet))
**Definition:** lmplot combines regression plot with faceting.
- Compare regression trends across categories

```python
sns.lmplot(
    data=df,
    x="Horsepower",
    y="Price",
    hue="Origin",
    height=5,
    aspect=1.2
)
plt.suptitle("LM Plot: Horsepower vs Price by Origin", y=1.02)
plt.show()
```
