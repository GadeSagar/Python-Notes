# What is Matplotlib?
- Matplotlib is a Python visualization library used to create static, animated, and interactive charts.
It is one of the most powerful plotting libraries and forms the foundation of other tools.

# Why is Matplotlib Used?

Matplotlib is used for:
1. Visualizing trends and patterns in data
2. Creating scientific and statistical charts
3. Building dashboards and reports
4. Data exploration during machine learning
5. Publishing-quality graphics for presentations
6. Customizing every single element of a chart

# Matplotlib Architecture (Important)

Matplotlib follows a layered architecture:

(1) Figure

The full window or page where all charts are drawn.

(2) Axes

The actual plotting area inside the figure.
Each subplot is one Axes.

(3) Axis

The x-axis and y-axis.

(4) Artist

Objects like:

    lines
    bars
    legends
    markers


```python
import pandas as pd
df1 = pd.read_csv('/content/50_Startups.csv')
df1.head(2)
```

# 1. LINE CHART
- A line chart is used to show the trend of a numeric variable over time or continuous changes.

### Purpose / When to Use

- To show growth, decline, or pattern changes
- To analyze time-series data
- To observe continuous variation

### Key Characteristics

- X-axis → Usually time
- Y-axis → Numeric values
- Uses continuous connected line segments

### Advantages

- Excellent for showing trends
- Easy to compare multiple lines
- Simple and readable

### Real-World Uses

- Sales trend over months
- Stock prices over time
- Temperature trend during a day

```python
# Line Plot.
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.plot(df1['RND'],df1['PROFIT'])
plt.title("RND vs Profit")
plt.xlabel("RND ")
plt.ylabel("PROFIT")
plt.grid()
plt.show()
```

# 2. BAR CHART (Vertical Bar)
- A bar chart uses vertical bars to compare values across different categories.

### Purpose / When to Use

- To compare categorical data
- To analyze highest, lowest, or differences between categories

### Key Characteristics

- Bars stand vertically
- Height = value
- Best for nominal or categorical data

### Advantages

- Easy to compare multiple categories
- Clear visual differences
- Works well with small and large categories

### Real-World Uses

- Profit by State
- Sales by Category
- Cars by Manufacturer count

```python
# Bar Chart - Average profit per state.
state_profit = df1.groupby('STATE')['PROFIT'].mean()

plt.figure(figsize=(7,5))
plt.bar(state_profit.index, state_profit.values)
plt.title("STATE VS PROFIT")
plt.xlabel("STATE ")
plt.ylabel("PROFIT")
plt.show()
```

# 3. SCATTER PLOT
- A scatter plot shows the relationship between two numeric variables using dots.

### Purpose / When to Use

- To identify correlation (positive/negative)
- To detect patterns, clusters, or outliers

### Key Characteristics

- Each point represents one observation
- X vs Y numeric comparison
- No lines connecting points

### Advantages

- Best for correlation analysis
- Highlights outliers clearly
- Shows data distribution effectively

### Real-World Uses

- R&D Spend vs Profit
- Age vs Salary
- Horsepower vs Mileage (Cars93)

```python
# Scatter Plot - Marketing vs PROFIT
# showing correlations.
# Identifying outliers.

plt.figure(figsize=(8,5))
plt.scatter(df1['MKT'], df1['PROFIT'])
plt.title("Marketing vs PROFIT")
plt.xlabel("Marketing")
plt.ylabel("PROFIT")
plt.show()
```

```python

plt.figure(figsize=(8,5))
plt.scatter(df1['RND'], df1['PROFIT'])
plt.title("RND vs PROFIT")
plt.xlabel("RND")
plt.ylabel("PROFIT")
plt.show()
```

# 4. HISTOGRAM
- A histogram shows the frequency distribution of a numeric variable by grouping data into bins.

### Purpose / When to Use

- To understand distribution
- To see skewness, spread, shape
- To find patterns in continuous numeric data

### Key Characteristics

- Bars touch each other
- X-axis contains bins (intervals)
- Y-axis shows frequency count

### Advantages

- Best for distribution analysis
- Detect outliers and skew
- Helps understand spread of data

### Real-World Uses

- Age distribution
- Salary distribution
- Car horsepower distribution

```python
# Histogram - Distribution of PROFIT.
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.hist(df1['PROFIT'],bins=10)
plt.title("Distribution of PROFIT")
plt.xlabel('PROFIT')
plt.ylabel("Count")
plt.show()
```

# 5. BOXPLOT
- A boxplot shows the spread, distribution, and outliers of a numeric dataset using five summary points:

            Minimum
            Q1 (25%)
            Median (50%)
            Q3 (75%)
            Maximum

### Purpose / When to Use

            To detect outliers
            To understand distribution
            To compare multiple groups

### Key Characteristics

            Box = IQR (Interquartile Range)
            Line inside box = median
            Whiskers = range
            Dots = outliers

### Advantages

            Best for comparing distributions
            Great for identifying extreme values

### Real-World Uses

            Profit distribution
            Salary comparison
            Engine size variation across car types

```python
# Boxplot - Profit for Outlier Detection.
# Finding outliers.
# Checking spread and median.
plt.figure(figsize=(6,5))
plt.boxplot(df1['PROFIT'])
plt.title("Profit Boxplot")
plt.ylabel('PROFIT')
plt.show()
```

# 7. PIE CHART
- A pie chart displays parts of a whole as slices of a circle.

### Purpose / When to Use

- To show percentage contribution
- To compare shares of a whole

### Key Characteristics

- 360° circle
- Each slice = category share
- Works best with limited categories (3–7)

### Advantages

- Good for quick percentage understanding
- Simple and visually appealing

### Real-World Uses

- Market share
- Sales percentage by region
- Car types proportion

```python
origin_counts = df['Type'].value_counts()

plt.figure(figsize=(7,7))
plt.pie(
    origin_counts,
    labels=origin_counts.index,
    autopct="%1.1f%%",
    startangle=140,
    shadow=True,
    colors=["gold", "lightblue", "lightgreen"]
)

plt.title("Cars90 Dataset: Car Origin Distribution")
plt.show()
```

# What Is a Subplot?

- A subplot means displaying multiple plots inside one single figure.

### It is helpful when you want to:

- Compare multiple charts side-by-side
- Show different variables in one view
- Create dashboards
- Visualize distributions, trends, and relationships together

### Why Use Subplots?

- Create dashboards
- Compare multiple variables
- Save space in reports
- Provide professional visualizations
- Show relationships and patterns in one figure

```python
import pandas as pd
df = pd.read_csv('/content/Cars93.csv')
df.head()
```

```python
# Subplot
plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
plt.scatter(x=df['Horsepower'], y=df['Price'])

plt.subplot(1,2,2)
plt.hist(df['Price'])

plt.show()
```

# LEGENDS
- Legend explains which color/line belongs to which label.

### Use When
- You have multiple lines or scatter groups
- Need to clarify categories

### Legend Position Options

    "upper left"
    "upper right"
    "lower left"
    "best" (automatic)


```python
from matplotlib.lines import lineStyles
plt.plot(df['MPG.city'], color = 'red',linestyle= ':', marker = '*')
plt.title('City MPG Trend')
plt.grid(False)
plt.show()
```

```python
# Two Line Charts

plt.figure(figsize=(10,5))
plt.plot(df['Price'], label = 'Price', color='blue')
plt.plot(df['MPG.city'], label='MPG City', color='Red')
plt.legend(loc="upper left")
plt.show()
```
